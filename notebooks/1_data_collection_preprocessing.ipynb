```python
df_temp = cleaned_df.copy()
df_temp.drop(columns=['tax_class_at_sale', 'building_class_at_sale'], inplace=True)
```

```python
df_temp['year_built'].fillna(0, inplace=True)  
df_temp['year_built'].replace([np.inf, -np.inf], 0, inplace=True)  

df_temp['apartment_number'] = df_temp['apartment_number'].apply(lambda x: x.strip() if isinstance(x, str) else x)
df_temp['apartment_number'] = df_temp['apartment_number'].replace('', np.nan)
1
df_temp['year_built'] = df_temp['year_built'].astype(int)
df_temp = df_temp[(df_temp['year_built'] > 0) & (df_temp['sale_price'] > 0) & (df_temp['land_square_feet'] > 0)]
df_temp.reset_index(drop=True, inplace=True)
df_temp['property_type'] = np.where(
    df_temp['apartment_number'].isna(), 
    np.where(df_temp['commercial_units'] >= 1, 'mixed house', 'house'),
    np.where(df_temp['commercial_units'] >= 1, 'mixed apartment', 'apartment')
)
df_temp['commercial_units'].fillna(0,inplace=True)

#density 
df_temp['land_density'] = df_temp['gross_square_feet'] / df_temp['land_square_feet']
df_temp['unit_density'] = df_temp['total_units'] / df_temp['gross_square_feet']

df_temp['log_sale_price'] = np.log1p(df_temp['sale_price'])
df_temp['log_land_square_feet'] = np.log1p(df_temp['land_square_feet'])
df_temp['log_gross_square_feet'] = np.log1p(df_temp['gross_square_feet'])

df_temp['res_com_interaction'] = df_temp['residential_units'] * df_temp['commercial_units']
df_temp['year_built_neighorhood_interaction'] = df_temp['year_built'] * df_temp['neighborhood'].factorize()[0]+1
```

```python
#density 
df_temp['land_density'] = df_temp['gross_square_feet'] / df_temp['land_square_feet']
df_temp['unit_density'] = df_temp['total_units'] / df_temp['gross_square_feet']

df_temp['log_sale_price'] = np.log1p(df_temp['sale_price'])
df_temp['log_land_square_feet'] = np.log1p(df_temp['land_square_feet'])
df_temp['log_gross_square_feet'] = np.log1p(df_temp['gross_square_feet'])

df_temp['res_com_interaction'] = df_temp['residential_units'] * df_temp['commercial_units']
df_temp['year_built_neighorhood_interaction'] = df_temp['year_built'] * df_temp['neighborhood'].factorize()[0]+1
```

```python
#Output cleaned dataset 
housing_df = df_temp.copy()
housing_df.to_csv('compiled_housing_sales.csv', index=False)
```
